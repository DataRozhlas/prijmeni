```python
import pandas as pd
import numpy as np
table = pd.read_csv("data.csv",index_col="PŘÍJMENÍ")
```

```python
# počet unikátních příjmení po letech
for year in range(1916,2017):
    print(np.sum(table[str(year)].values!=0),end=",")
```

```python
# počet "osamělých" příjmení
filtable = table[table[str(3000)] == 1]

for year in range(1916,2017):
    print(np.sum(filtable[str(year)].values==1),end=",")
```

```python
filsubtable = filtable[filtable[str(1970)] == 1]
filsubtable.filter(items=["PŘÍJMENÍ", "1970"])
```

```python
# mediánový věk pro jméno
def get_median(name):
    subtable = table.loc[name].loc["1916":"2016"]
    year = 1916
    median_list = []
    for entry in subtable:
        for i in range(int(entry)):
            median_list.append(int(year))
        year+=1

    return median_list[round(len(median_list)/4)], median_list[round(len(median_list)/2)], median_list[3*round(len(median_list)/4)]
```

```python
namecount = 0
name_medians = {}
for name, row in table.iterrows():
    if row["3000"] >= 1000:
        namecount += 1
        name_medians[name] = get_median(name)
print(namecount)
```

```python
medians_sorted = sorted(name_medians, key=lambda x: name_medians[x][1])

# nejstarší příjmení (podle medianu)
for name in medians_sorted[:120]:
    print("[" + str(2017-name_medians[name][2]) + ", " +str(2017-name_medians[name][0]) + "], //" + name.title(), end=" ")
    print(2017-name_medians[name][1])
```

```python
# nejmladší jména (podle medianu)
medians_sorted = sorted(name_medians, key=lambda x: name_medians[x][1], reverse=True)
for name in medians_sorted[:120]:
    if name != "NEZJIŠTĚNO":
        print("[" + str(2017-name_medians[name][2]) + ", " +str(2017-name_medians[name][0]) + "], //" + name.title(), end=" ")
        print(2017-name_medians[name][1])
```

```python
# největší spread
spread = {}
for name in name_medians:
    spread[name] = int(name_medians[name][2])-int(name_medians[name][0])
spread = sorted(spread, key=lambda x: int(spread[x]),reverse=True)
for name in spread[:25]:
        print("[" + str(2017-name_medians[name][2]) + ", " +str(2017-name_medians[name][0]) + "], //" + name.title(), end=" ")
        print(2017-name_medians[name][1])
for name in spread[:25]:
    print('"' + name.title() + '", ',end="")
for name in spread[:25]:
    print(str(2017-name_medians[name][1]) + ', ',end="")
```
